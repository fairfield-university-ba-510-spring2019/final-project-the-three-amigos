**We want to make sure that Anita's information is still correct from the data warehouse. She does have the same instructor ID from the database.**

```python
%%sql

SELECT IID, Name FROM INSTRUCTOR_WH WHERE IID = 55;
```

**Quickly, we want to see the Title of some courses, and if there are any classes that have different locations and meeting times.**

```python
%%sql

SELECT Title, LID, IID, MID
FROM COURSES_WH
WHERE IID = 55;
```

**Our next check is to make sure that the query from our database works for Anita, and it does. Our data warehouse is telling us that Anita has taught 38 distinct classes.**

```python
%%sql

SELECT COUNT (DISTINCT CRN) AS Classes FROM COURSES_WH WHERE IID = 55;
```

**Theses are the amount of classes that Anita teaches based on the title of the class. Some of the titles may be slightly different, such as the 'Internship' title, as there is an 's' in one of them, and a '-' in the other. That seems to be a bug for the catalog/course, since they have the same serial number of 201409.**

```python
%%sql

SELECT DISTINCT(Title), COUNT(Title) AS Count FROM COURSES_WH WHERE IID = 55 GROUP BY Title;
```

**Now we want to see which locations Anita is teaching in. This _DOES NOT_ represent the total amount of class sessions that she has taught in like the database, but the amount of different locations each of her classes have been taught in.**

```python
%%sql

SELECT LID, COUNT(*) AS Count
FROM COURSES_WH
WHERE IID = 55
GROUP BY LID
ORDER BY Count DESC;
```
